# 前処理

```python
import pandas as pd
#データ取得
df_orig = pd.read_csv("../../data/train.csv").drop("Unnamed: 0", axis=1).dropna()
#データ分割
df_eco = df_orig.query('Class == "Eco"')
df_eco_plus = df_orig.query('Class == "Eco Plus"')
df_business= df_orig.query('Class == "Business"')
#データ数削減
df_eco_new = df_eco.sample(n=11500, random_state=42)
df_business_new = df_business.sample(n=11500, random_state=42)
#データを結合
df = pd.concat([df_eco_new, df_business_new, df_eco_plus], axis=0)
#クラスの割合を確認
df["Class"].value_counts()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
df.head()
```

# 調査
全席クラスに対して、クラスタリングせずに、各満足度を可視化、調査

```python
df.columns
```

```python
#Inflight wifi service
feat="Inflight wifi service"
cross_count = pd.crosstab(df[feat],
                    df["satisfaction"], margins=True)
cross_proportion = pd.crosstab(df[feat],
                    df["satisfaction"], margins=True, normalize="index")

fig = plt.figure(121, figsize=(15,5))
ax_count = fig.add_subplot(122)
ax_proportion = fig.add_subplot(121)

cross_count.plot(kind="bar", ax=ax_count)
ax_count.set_title("count")
cross_proportion.plot(kind="bar", ax=ax_proportion)
ax_proportion.set_title("proportion")

plt.show()
```

データ全体に対してもInflight wifi serviceとtotal satisfactionに正の関係がある。\
一般的にInflight wifi serviceには力を入れたほうがいいかもしれない

# 各満足度の平均点
各満足度の平均を算出してデータフレームに追加

```python
#各満足度の列名取得
cols_respective_satisfactions = df.columns[7:21]
cols_respective_satisfactions
```

```python
#各満足度の平均を特徴量として追加
df["mean_satisfactions"] = df[cols_respective_satisfactions].mean(axis=1)
df.head()
```

```python
#mean_satisfactions
feat = "mean_satisfactions"
fig = plt.figure(figsize=(6,4))
ax_all = fig.add_subplot(111)

sns.histplot(data=df, x=feat, hue="satisfaction", kde=True, bins=15, ax=ax_all)
ax_all.set_title("All Data histogram")

plt.show()
```

予想通り二峰性のヒストグラムになった\
・平均点が低いのにtotalで満足している人\
・平均点が高いのにtotalで満足していない人\
この2つがキーかもしれない

```python
#各満足度の平均を特徴量として追加
satisfied_mean = pd.DataFrame(df.query('satisfaction=="satisfied"')[cols_respective_satisfactions].mean(axis=0),
                              columns=["Mean"])
satisfied_mean["satisfaction"]="satisfied"
satisfied_mean
```

```python
#各満足度の平均を特徴量として追加
not_satisfied_mean = pd.DataFrame(df.query('satisfaction!="satisfied"')[cols_respective_satisfactions].mean(axis=0),
                              columns=["Mean"])
not_satisfied_mean["satisfaction"]="not_satisfied"
not_satisfied_mean
```

```python
respective_satisfaction_mean = pd.concat([satisfied_mean, not_satisfied_mean])
#respective_satisfaction_mean.columns=["satisfied", "not_satisfied"]
respective_satisfaction_mean
```

```python
#各満足度の平均を特徴量として追加
sns.barplot(data=respective_satisfaction_mean,
            x="Mean",
            y=respective_satisfaction_mean.index,
               hue="satisfaction")
plt.title("Mean of Respective Satisfactions from all data ")
plt.ylabel("Respective Satisfactions")
plt.xlabel("Mean of Satisfaction")
plt.show()
```

```python
#各満足度の平均を特徴量として追加
sns.barplot(x=df[cols_respective_satisfactions].mean(axis=0).sort_values(),
        y=df[cols_respective_satisfactions].mean(axis=0).sort_values().index,
           orient="h")
plt.title("Mean of Respective Satisfactions from all data ")
plt.ylabel("Respective Satisfactions")
plt.xlabel("Mean of Satisfaction")
plt.show()
```
