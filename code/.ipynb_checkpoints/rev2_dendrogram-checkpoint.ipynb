# デンドログラムでクラスタリング

## 前処理

```python
import pandas as pd
#データ取得
df_orig = pd.read_csv("../data/train.csv").drop("Unnamed: 0", axis=1).dropna()
#データ分割
df_eco = df_orig.query('Class == "Eco"')
df_eco_plus = df_orig.query('Class == "Eco Plus"')
df_business= df_orig.query('Class == "Business"')
#データ数削減
df_eco_new = df_eco.sample(n=11500, random_state=42)
df_business_new = df_business.sample(n=11500, random_state=42)
#データを結合
df = pd.concat([df_eco_new, df_business_new, df_eco_plus], axis=0)
#クラスの割合を確認
df["Class"].value_counts()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
df.columns
```

```python
cluster_features=['Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class',
       'Flight Distance','Departure Delay in Minutes', 'Arrival Delay in Minutes']
```

```python
#デンドログラムを実施するdfを作成
df_dend = df[cluster_features]
df_dend.head(3)
```

```python
#ダミー変数化
df_dummies = pd.get_dummies(df_dend, drop_first=True, dtype="int")
df_dummies.head(3)
```

```python
#標準化
ss = StandardScaler()
df_ss = pd.DataFrame(ss.fit_transform(df_dummies), columns=df_dummies.columns)
df_ss.head(3)
```

## クラスタリング

```python
#デンドログラム
result = linkage(df_ss, method="ward")
dendrogram(result)
plt.show()
```

```python
#クラスタラベル取得
from scipy.cluster.hierarchy import fcluster
cluster = fcluster(result, t=5, criterion = "maxclust")
cluster
```

```python
df["cluster"] = cluster.astype("str")
df.head()
```

# 調査

## クラスタごとのの満足度

```python

feat="cluster"
cross_count = pd.crosstab(df[feat],
                    df["satisfaction"], margins=True)
cross_proportion = pd.crosstab(df[feat],
                    df["satisfaction"], margins=True, normalize="index")

fig = plt.figure(121, figsize=(15,5))
ax_count = fig.add_subplot(122)
ax_proportion = fig.add_subplot(121)

cross_count.plot(kind="bar", ax=ax_count)
ax_count.set_title("count")
cross_proportion.plot(kind="bar", ax=ax_proportion)
ax_proportion.set_title("proportion")

plt.show()
```

```python
#クラスタ4の全体に対する割合
100*len(df.query('cluster=="4"'))/len(df)
```

```python
#クラスタ4の不満足者の全体に対する割合
100*len(df.query('cluster=="4" & satisfaction=="neutral or dissatisfied"'))/len(df)
```

```python
#満足していない人の割合
100*len(df.query("satisfaction!='satisfied'"))/len(df)
```
