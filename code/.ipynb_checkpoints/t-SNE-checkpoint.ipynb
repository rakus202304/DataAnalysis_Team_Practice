```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#データ読み込み
df = pd.read_csv("./data/train.csv")
df.head(3)
```

```python
#欠損値の削除
df = df.dropna()
#id,unnamed 不要な列を消す
df_deleted = df.drop(['id','Unnamed: 0'], axis=1)
# One-hot encodingを実行する
encoded_df = pd.get_dummies((df_deleted), dtype=int,drop_first=True)
encoded_df.head()
```

```python
#eco_plusのデータを抽出
eco_plus = encoded_df[encoded_df['Class_Eco Plus']==1]\
    .drop(["Class_Eco", "Class_Eco Plus", "satisfaction_satisfied"], axis=1)#Classに関する列を落とす
#satisfactionを取得
satisfy = encoded_df[encoded_df['Class_Eco Plus']==1].satisfaction_satisfied
```

```python
eco_plus.shape
```

## t-SNE

```python
from sklearn.manifold import TSNE
#インスタンス化
tsne = TSNE(n_components = 2,
            perplexity = 50.0,
            early_exaggeration = 12.0,
            learning_rate = 400.0,
            method = "barnes_hut",
            random_state = 42)
```

```python
#t-SNEの実行
X = tsne.fit_transform(eco_plus)
X.shape
```

```python
#satisfyを追加して、データフレーム化
tsne_df = pd.DataFrame([X[:,0], X[:,1], satisfy.values]).T
tsne_df.columns= ["column1", "column2", "satisfy"]
tsne_df.head()
```

```python
#可視化
import seaborn as sns
sns.scatterplot(data=tsne_df, x="column1", y="column2", hue="satisfy")
plt.plot()
```

```python
from sklearn.manifold import TSNE
#インスタンス化
tsne = TSNE(n_components = 2,
            perplexity = 20.0,
            early_exaggeration = 50,
            learning_rate = 400.0,
            method = "barnes_hut",
            random_state = 42)

#t-SNEの実行
X = tsne.fit_transform(eco_plus)

#satisfyを追加して、データフレーム化
tsne_df = pd.DataFrame([X[:,0], X[:,1], satisfy.values]).T
tsne_df.columns= ["column1", "column2", "satisfy"]

#可視化
import seaborn as sns
sns.scatterplot(data=tsne_df, x="column1", y="column2", hue="satisfy")
plt.gca().axis('off')
#plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.plot()
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(eco_plus)
X_pca
```

```python
#satisfyを追加して、データフレーム化
pca_df = pd.DataFrame([X_pca[:,0], X_pca[:,1], satisfy.values]).T
pca_df.columns= ["column1", "column2", "satisfy"]

#可視化
import seaborn as sns
sns.scatterplot(data=pca_df, x="column1", y="column2", hue="satisfy")
plt.gca().axis('off')
#plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
plt.plot()
```

```python
sns.displot(pca_df, x="column1", hue="satisfy")
plt.plot()
```
