```python
class process(object):
    """
    def __init__(self, df):
        self.df = df
    """
    
    # ダミー変数化
    def one_hot_encoding(self, df):
        df = pd.get_dummies(df, drop_first=False, dtype="int")
        return df

    #データ分割
    def data_split(self, df):
        X = df.drop("HeartDisease", axis=1)
        y = df.HeartDisease
        return X, y

    def data_replace(self, df):
        #コレステロールの値を置換
        df.loc[df.query('Cholesterol==0 and Sex==0 and FastingBS==0').index,"Cholesterol"] = 242
        df.loc[df.query('Cholesterol==0 and Sex==0 and FastingBS==1').index,"Cholesterol"] = 288
        df.loc[df.query('Cholesterol==0 and Sex==1 and FastingBS==0').index,"Cholesterol"] = 231
        df.loc[df.query('Cholesterol==0 and Sex==1 and FastingBS==1').index,"Cholesterol"] = 240
        return df

    #標準化
    def Do_ss(self, base_df, main_df):
        ss = StandardScaler()
        ss.fit(base_df)
        main_df = pd.DataFrame(ss.transform(main_df), columns=base_df.columns)
        return main_df
```

```python
#インスタンス化
cl_process = process()
#コレステロール値を置換
raw_df = cl_process.data_replace(raw_df)
raw_df.head(3)
```

```python
sns.displot(data=raw_df, x="Cholesterol", hue="HeartDisease", kde=True, bins=15)
plt.show()
```

```python
cross = pd.crosstab(raw_df.FastingBS, raw_df.HeartDisease, margins=True, normalize="index")
display(cross)
cross.plot(kind="bar")
plt.show()
```

FastingBS=1を糖尿傾向として、糖尿傾向の方が心臓病になりやすい\
ただし、糖尿傾向じゃない場合も50%弱の割合で心臓病がいるから、逆は言えない\
